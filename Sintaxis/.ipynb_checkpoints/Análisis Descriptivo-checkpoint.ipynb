```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 5)
```

### Leer los datos

```python
tesis = pd.read_csv('../Datos/tesis_final.csv')
```

```python
tesis.head()
```

```python
numerical = ['Peso', 'Altura', 'Cintura', 'Cadera', 'IMC', 'Edad', 'tas_basal']
categorical = ['IMC_30', 'DBT', 'Tabaquismo', 'Sexo', 'Fuma', 'ant_HTA', 'ant_HTA_trat', 'n_drogas_actual_categ', 'factor_riesgo']
```

### Análisis de la respuesta

```python
# Scatter plot
sns.scatterplot(x=tesis['tpo_programa'], y=tesis['TAS'], alpha=0.2)
```

```python
def jitter(arr):
    stdev = .01*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev
```

```python
# Mismo gráfico pero con jitter para ver mejor la concentración de los datos
sns.scatterplot(x=jitter(tesis['tpo_programa']), y=jitter(tesis['TAS']), alpha=0.2)
```

```python
# Scatterplot con regresión (no lo hice todo junto porque no tiene parametro alpha)
sns.regplot(x=tesis['tpo_programa'], y=tesis['TAS'], scatter_kws={'color': 'steelblue'}, line_kws={'color': 'red'})
```

```python
# Mismo gráfico anterior pero con jitter y loess
sns.regplot(x=jitter(tesis['tpo_programa']), y=jitter(tesis['TAS']), scatter_kws={'color': 'steelblue'}, line_kws={'color': 'red'}, lowess=0.5)
```

```python
# Gráfico de perfiles promedio
sns.lineplot(x=tesis['tpo_programa'], y=tesis['TAS'], marker='o')
```

### Análisis descriptivo univariado

```python
tesis_basal = tesis[tesis['tpo_programa'] == 0]
```

```python
# Variable respuesta
sns.distplot(tesis_basal['tas_basal'])
```

```python
# Plot features
plt.figure(figsize=(15,15))

for i, x in enumerate(numerical):
    axs = plt.subplot((len(numerical)/2)+1,2,i+1)
    dist = sns.distplot(tesis_basal[x])
    dist.axes.set_xlabel(str(x))

plt.tight_layout()
plt.show()
```

```python
for x in categorical:
    print('*'*50)
    print(tesis_basal[x].value_counts(normalize=True))
```

### Análisis descriptivo bivariado

```python
# Covariables numéricas
plt.figure(figsize=(15,15))

for i, x in enumerate(numerical):
    axs = plt.subplot((len(numerical)/2)+1,2,i+1)
    sns.regplot(x=tesis_basal[x], y=tesis_basal['TAS'], scatter_kws={'color': 'steelblue'}, line_kws={'color': 'red'})

plt.tight_layout()
plt.show()
```

```python
# Covariables categóricas
plt.figure(figsize=(15,15))

for i, x in enumerate(categorical):
    tesis_not_null = tesis_basal[tesis_basal[x].notnull()]
    axs = plt.subplot((len(categorical)/2)+1,2,i+1)
    for value in tesis_not_null[x].unique():
        dist = sns.distplot(tesis_not_null.loc[tesis_not_null[x] == value,'TAS'], label=value)
    dist.axes.set_xlabel('TAS')
    plt.legend()
    plt.title(str(x))

plt.tight_layout()
plt.show()
```

### Análisis bivariado a través del tiempo

```python
# Covariables numéricas
plt.figure(figsize=(15,15))

for i, x in enumerate(numerical):
    axs = plt.subplot((len(numerical)/2)+1,2,i+1)
    line = sns.lineplot(x=tesis['tpo_programa'], y=tesis['TAS'], hue=tesis[x] < tesis[x].median())
    plt.legend()
    plt.title(str(x))

plt.tight_layout()
plt.show()
```

```python
# Covariables numéricas
plt.figure(figsize=(15,15))

for i, x in enumerate(categorical):
    tesis_aux = tesis[tesis[x].notnull()]
    axs = plt.subplot((len(categorical)/2)+1,2,i+1)
    line = sns.lineplot(x=tesis_aux['tpo_programa'], y=tesis_aux['TAS'], hue=tesis_aux[x].astype('int64'))
    plt.legend()
    plt.title(str(x))

plt.tight_layout()
plt.show()
```

### Análisis de Morisky

```python
sns.lineplot(x=tesis['tpo_programa'], y=tesis['TAS'], hue=tesis['Morisky'])
```

```python
# Porcentaje de adherencia a través del tiempo
sns.lineplot(x=tesis['tpo_programa'], y=tesis['Morisky'])
```
