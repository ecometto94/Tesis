```python
import pandas as pd
import numpy as np
```

#### Leer los datos

```python
tesis = pd.read_csv("../Datos/Datos_Tesis_Septiembre2020.csv", na_values=' ', index_col=0)
```

```python
tesis.rename({'TAS_noe':'TAS'}, axis=1, inplace=True)
```

```python
tesis.head()
```

```python
# Remover variables que no voy a usar para leer mejor los datos
tesis.drop(['DSCVisita', 'fechavisita', 'fecha_visitabasal', 'IMC_25', 'ICC_alto'], axis=1, inplace=True)
```

#### Info de las covariables

```python
tesis.info()
```

Todas las variables se leyeron como numéricas, pero hay muchas que son categóricas asique tengo que cambiarle el tipo de variable

```python
categorical = ['idPaciente', 'TA_ALTA', 'Morisky', 'IMC_30', 'DBT', 'Tabaquismo', 'Sexo', 'Fuma', 'ant_HTA', 'ant_HTA_trat', 'n_drogas_actual_categ']
numerical = ['tpo_programa', 'TAS', 'Peso', 'Altura', 'Cintura', 'Cadera', 'IMC', 'Edad']

for x in categorical:
    tesis[x] = tesis[x].astype('object')
```

#### Análisis de valores faltantes

```python
round(tesis.isnull().sum()/len(tesis)*100, 2)
```

#### Estadísticas básicas covariables continuas

```python
tesis.describe()
```

Altura, Cintura y Cadera tienen de valor mínimo 0, lo cual es imposible, por lo tanto voy a imputarlos con la mediana

```python
for x in ['Altura', 'Cintura', 'Cadera']:
    tesis.loc[tesis[x] == 0, x] = np.nan
```

```python
tesis.describe()
```

#### Análisis covariables categóricas

```python
for x in categorical:
    if x != 'idPaciente':
        print('*'*50)
        print(round(tesis[x].value_counts(normalize=True)*100, 2))
```

Sexo tiene valores 1 y 2, tengo que pasarlos a 0 y 1 <br>
hay solo 0.09% de pacientes que toman 0 drogas, voy a juntarlos con los que toman 1

```python
tesis.loc[tesis['Sexo'] == 2, 'Sexo'] = 0
tesis.loc[tesis['n_drogas_actual_categ'] == 0, 'n_drogas_actual_categ'] = 1
```

```python
for x in ['Sexo', 'n_drogas_actual_categ']:
    print('*'*50)
    print(round(tesis[x].value_counts(normalize=True)*100, 2))
```

#### Crear covariable Factor de Riesgo

```python
fr = []
for i in tesis.index:
    if tesis['DBT'][i] == 1 or tesis['Fuma'][i] == 1 or tesis['IMC_30'][i] == 1:
        fr.append('1')
    else:
        fr.append('0')
        
tesis['factor_riesgo'] = fr
tesis.head()
```

```python
tesis.groupby('factor_riesgo').idPaciente.nunique()
```

#### Tratamiento de la variable Morisky

```python
# Crear morisky_acum: variable de indice de performance de adherencia al tratamiento hasta el momento t
morisky_acum = []
divisor = 0
morisky_acum_aux = 0
idPaciente_aux = tesis['idPaciente'][0]
for i in tesis.index:
    if idPaciente_aux == tesis['idPaciente'][i]:
        divisor += 1
        morisky_acum_aux += tesis['Morisky'][i]
        morisky_acum.append(morisky_acum_aux/divisor)
        idPaciente_aux = tesis['idPaciente'][i]
    else:
        divisor = 1
        morisky_acum_aux = tesis['Morisky'][i]
        morisky_acum.append(morisky_acum_aux/divisor)
        idPaciente_aux = tesis['idPaciente'][i]
        
tesis['Morisky_Acum'] = morisky_acum
tesis.head()
```

```python
# Pasar la covariable a numérica para poder calcularle la media
tesis['Morisky'] = tesis['Morisky'].astype('int64')

# Diccionario con performance de morisky por idpaciente
perf_dict = dict(tesis.groupby('idPaciente').Morisky.mean())

# Crear covariable Morisky_Perf no dependiente del tiempo con el performance final de cada paciente
morisky_perf = []
for i in tesis.index:
    morisky_perf.append(perf_dict[tesis['idPaciente'][i]])
    
tesis['Morisky_Perf'] = morisky_perf
tesis.head()
```

#### Crear covariable TAS_media_acum (TAS hasta el tiempo actual)

```python
tas_media_acum = []
divisor = 0
tas_acum_aux = 0
idPaciente_aux = tesis['idPaciente'][0]
for i in tesis.index:
    if idPaciente_aux == tesis['idPaciente'][i]:
        divisor += 1
        tas_acum_aux += tesis['TAS'][i]
        tas_media_acum.append(tas_acum_aux/divisor)
        idPaciente_aux = tesis['idPaciente'][i]
    else:
        divisor = 1
        tas_acum_aux = tesis['Morisky'][i]
        tas_media_acum.append(tas_acum_aux/divisor)
        idPaciente_aux = tesis['idPaciente'][i]
        
tesis['TAS_media_acum'] = tas_media_acum
tesis.head()
```

#### Guardar dataset limpio

```python
tesis.to_csv('../Datos/tesis_final.csv', index=False)
```
